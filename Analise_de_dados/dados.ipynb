```python
# Importando a biblioteca pandas para manipulação de dados
import pandas as pd

banco_dados = pd.read_csv('banco_clientes_ficticios_v4.csv')
display(banco_dados)
```

```python
# Verificando a quantidade de valores nulos em cada coluna

display(banco_dados.isnull().sum())
```

```python
# Contar a quantidade de clientes por Status, Plano e Estado
display(banco_dados['Status'].value_counts())
display(banco_dados['Plano'].value_counts())
display(banco_dados['Estado'].value_counts())
```

```python
# Alterando o culuna de preço para o tipo numérico, removendo o símbolo de moeda e convertendo para float

banco_dados['Preço'] = banco_dados['Preço'].str.replace('R$', '').str.replace(',', '.').astype(float)
banco_dados = banco_dados.dropna().drop_duplicates()



display(banco_dados)
```

```python
#Convertendo Data Assinatura para o tipo data
banco_dados["Data Assinatura"] = pd.to_datetime(banco_dados["Data Assinatura"], errors="coerce")
display(banco_dados)
```

```python
# Calculando a Idade media dos clientes
idade_media = banco_dados['Idade'].mean()
print(f'Idade média dos clientes: {idade_media:.1f} anos')
```

```python
# Calculando idade média por plano
idade_media_plano = banco_dados.groupby('Plano')['Idade'].mean()
print(f'Idade média por plano:\n{idade_media_plano} anos')
```

```python
# Receita total por plano
receita_total_plano = banco_dados.groupby('Plano')['Preço'].sum()
print(f'Receita total por plano:\n{receita_total_plano.round(2)}')
```
